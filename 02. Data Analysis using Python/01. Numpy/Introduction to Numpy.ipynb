#### What is Numpy?

1. Numpy Stands for Numerical Python
2. Numpy is the Package that defines a multi- dimensional array object,various derived objects and associated fast math functions that operates on it.
3. It also has functions for working in domain of linear algebra, Fourier Transformation and Matrics.
4. Its is fundamental package for Scientific Computing in Python

#### What are Array ?
1. An array is define as collection of items that are sorted at contiguoug memory  locations.
2. It is a container which can hold fixed number of items and this item should be of same datatype.


#### Advantages of using Numpy.
1. Numpy uses much less memory to store data.
2. Numpy make it extremely easy to perform Mathematical operations on it.
3. It is for creation of N-Dimensional array.
4. Finding elements in Numpy array is easy.
