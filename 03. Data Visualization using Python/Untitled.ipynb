


## 🎯 **What is Skewness?**

**Skewness** measures the asymmetry of the distribution of data.

* If data is symmetric, it’s normally distributed.
* If one tail is longer or fatter, the data is skewed.

---

## 🔹 **1. Symmetric Distribution (Normal Distribution)**

### ✅ **Definition:**

* The left and right sides of the histogram are mirror images.
* Most values cluster around the center.

### 📊 **Central Tendency Relationship:**

```
Mean = Median = Mode
```

### 📈 **Histogram Shape:**

* Bell-shaped curve.
* Perfectly balanced on both sides.

### 💡 **Examples:**

* Heights of adult humans.
* IQ scores.
* Measurement errors in scientific experiments.

---

## 🔸 **2. Right-Skewed Distribution (Positive Skew)**

### ✅ **Definition:**

* Tail extends to the **right** (toward higher values).
* A few extreme **high values** pull the mean to the right.

### 📊 **Central Tendency Relationship:**

```
Mode < Median < Mean
```

### 📈 **Histogram Shape:**

* Peak on the left.
* Long tail on the right.

### 💡 **Examples:**

* Income distribution (few people earn very high).
* House prices in a mixed neighborhood.
* Hospital stay durations.

---

## 🔸 **3. Left-Skewed Distribution (Negative Skew)**

### ✅ **Definition:**

* Tail extends to the **left** (toward lower values).
* A few extreme **low values** pull the mean to the left.

### 📊 **Central Tendency Relationship:**

```
Mean < Median < Mode
```

### 📈 **Histogram Shape:**

* Peak on the right.
* Long tail on the left.

### 💡 **Examples:**

* Age at retirement (most retire at 60+, but few early retire at 30–40).
* Exam scores when the test is too easy (most score high, few score very low).
* Time to solve a very easy puzzle (most are fast, few are slow).

---

## 🧠 **Quick Mnemonics:**

| Type           | Tail Direction | Central Tendencies   |
| -------------- | -------------- | -------------------- |
| Symmetric      | None           | Mean = Median = Mode |
| Right Skew (+) | Tail to Right  | Mode < Median < Mean |
| Left Skew (–)  | Tail to Left   | Mean < Median < Mode |
