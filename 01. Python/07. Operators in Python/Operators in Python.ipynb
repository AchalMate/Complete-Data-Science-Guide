### Python Operators
* Operators are used to perform operations on variables and values.
* Python divides the operators in the following groups:

    1. Arithmetic operators
    2. Assignment operators
    3. Comparison operators
    4. Logical operators
    5. Identity operators
    6. Membership operators
    7. Bitwise operators

### 1. Python Arithmetic Operators

* Arithmetic operators are used to on numeric values to perform basic mathematical operations.
* Arithmetic Operators are used between two operands



#### Operation on Numeric values

```python
a  = 10
b =  20
print('Addition of two numbers:',a+b)  
print('Subtraction of two numbers:',a-b)  
print('Multiplication of two numbers:',a*b)  
print('Division of two numbers:',a/b)  
print('Reminder of two numbers:',a%b)  
print('Exponent of two numbers:',a**b)  
print('Floor division of two numbers:',a//b)  
```

### Operation on list and tuple

```python
l1 = [1,2,3]
l2 = [2,3,4]
l1+l2
```

#### Operation on String Value

```python
x = "Hello"
y = "world"
print("Addition Operator works as concation function",x+y)
```

### 2. Assignment Operator
* Assignment operators are used to assign values to variables.
* The right expression's value is assigned to the left operand



```python
a = 32         # Initialize the value of a  
b = 6          # Initialize the value of b  
print('a=b:', a==b)  
print('a+=b:', a+b)  
print('a-=b:', a-b)  
print('a*=b:', a*b)  
print('a%=b:', a%b)  
print('a**=b:', a**b)  
print('a//=b:', a//b)  
```

### 3.Python Comparison Operators
Comparison operators compare the values of the two operands and return a true or false Boolean value in accordance.


```python
a = 5
b = 3

print(a == b)  # False
print(a != b)  # True
print(a > b)   # True
print(a < b)   # False
print(a >= b)  # True
print(a <= b)  # False
```

### 4.Logical Operator
* Logical operators are used to combine conditional statements



```python
a = 5          # initialize the value of a          
print('Is this statement true?:',a > 3 and a < 5)  
print('Any one statement is true?:',a > 3 or a < 5)  
print('Each statement is true then return False and vice-versa:',(not(a > 3 and a < 5)))  
```

### 5. Identity Operators
* Used to compare the memory locations of two objects.
* Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:



```python
a = ["Rose", "Lotus"]  
b = ["Rose", "Lotus"]  
c = a  
print("a is c",a is c)  
print('a is not c',a is not c)  
print('a is b',a is b)  
print('a is not b',a is not b)  
print('a == b',a == b)  
print('a != b',a != b)  
```

### 6. Membership operators
* Membership operators are used to test if a sequence is presented in an object:



```python
lst = [1, 2, 3, 4]

print(3 in lst)      # True
print(5 not in lst)  # True
```

```python
x = ["Rose", "Lotus"]  
print(' Is value Present?', "Rose" in x)  
print(' Is value not Present?', "Riya" not in x)  
```

7. Bitwise operators
* Bitwise operators are used to compare (binary) numbers:
* Bitwise operators perform bit-by-bit operations on integer values



```python
a = 5 # 101 in binary
b = 3  # 011 in binary

print(a & b)  # 1 (001)
print(a | b)  # 7 (111)
print(a ^ b)  # 6 (110)
print(~a)     # -6 (two's complement)
print(b << 1) # 10 (1010)
print(a >> 1) # 2 (010)
```

### Operator Precedence
* Operator precedence describes the order in which operations are performed.
* The precedence order is described in the table below, starting with the highest precedence at the top:


```python
# Parentheses has the highest precedence, meaning that expressions inside parentheses must be evaluated first:

print((6 + 3) - (6 + 3))
```

```python
# Multiplication * has higher precedence than addition +, and therefor multiplications are evaluated before additions:

print(100 + 5 * 3)
```
