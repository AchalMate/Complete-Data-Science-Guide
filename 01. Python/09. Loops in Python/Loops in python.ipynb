### Looping Statements (Iteration)
* Loops allow you to repeat a block of code multiple times. 
* Python provides two types of loops: 
    1. for 
    2. while

### 1. For Loop
* The For Loops in Python are a special type of loop statement that is used for sequential traversal. 
* The for loop iterates over a sequence (such as a list, tuple, string Dictionary,Set) and executes the block of code for each element in the sequence.

### Flow chart


Syntax
for var in iterable:
    # statements
    pass

### Python For Loop with String

```python
# Iterating over a String
s = "Hello World"
for i in s:
    # print(i)

    print(i,end = "")
```

### For Loop to travel a list

```python
list1 = [2,3,5,7,1,8,0,9]
for item in list1:
    print(item)
```

### Python for loop with Range

*  range(starting, ending,skipping)
    * Ex 1 to 10
    * starting = 1 (starting position value is included)
    * ending = 10+1 (ending position value is excluded)
    * skipping = 1/2/3


```python
for i in range(0, 10, 2):
    print(i)
```

```python
l1 = ["eat", "sleep", "repeat"]
for i in range(len(l1)):
    print(l1[i])
```

```python
for i in l1:
    print(i)
```

```python
for count, ele in enumerate(l1):
    print (count, ele)
```

#### Travel a tuple using for loop

```python
tp = (2,4,6)
for i in tp:
    print(i)
```

```python
t = ((1, 2), (3, 4), (5, 6))
for a,b in t:
    print(a,b)
```

```python
student_info = (("Alice", 23), ("Bob", 20), ("Charlie", 25))
 
for name, age in student_info:
    print(f"{name} is {age} years old.")
```

### set

```python
s1 = {2,4,6}
for i in s1:
    print(i)
```

### Dictionary

```python
d1 = {1:"one",2:"two"}
for i,j in d1.items():
    print(i,j)
```

```python
d = dict()

d['xyz'] = 123
d['abc'] = 345
for i in d:
    print("% s % d" % (i, d[i]))
```

```python
fruits = ["apple", "banana", "cherry"]
colors = ["red", "yellow", "green"]
for fruit, color in zip(fruits, colors):
    print(fruit, "is", color)
```

### Even/Odd Number logic

```python
for i in range(30,41):
    if i % 2 == 0:
        print("Even Number",i)
    else :
        print("Odd Number",i)
```

```python
l2 = [65,98,55]
for i in l2:
    if i % 2 == 0:
        print("Even Number",i)
    else :
        print("Odd Number",i)
```

### 2. While loop

* Python While Loop is used to execute a block of statements repeatedly until a given condition is satisfied/True. 
* When the condition becomes false, the line immediately after the loop in the program is executed.

Flow chart



```python
count = 1
while count <= 5:
    print(count)
    count += 1
```
